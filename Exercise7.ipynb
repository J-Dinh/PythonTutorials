Plot to show Anthony Fauci

```python
sns.set_style("darkgrid")
sns.displot(data, x="Mutation_rate", hue="Strain", kind="kde", fill=True, alpha=.4, legend=False)
plt.xlabel("Mutation Rate")
plt.title("Mutation Rates for 4 Virus Strains")
plt.xticks(np.arange(0, 20, 2))
plt.legend(["Delta", "Gamma", "Beta", "Alpha"], title="Strains")
```

The KDE plot shows that each strain has a different mutation rate. The Alpha strain has a large range of mutation rates with the average being around 10 mutations per generation x 10e-5. The Beta strain has a slightly smaller range but more desnity for around 5 mutations. The Delta strain has little range but high density for 2 mutations. The Gamma range has the smallest range and highest density for 1 mutation.
